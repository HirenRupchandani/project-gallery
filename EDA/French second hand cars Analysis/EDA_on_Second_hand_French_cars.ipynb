```python
train.columns
```

```python
numerical_cols = ['année', 'nombredeportes', 'nombredeplaces', 'critair', 'departement','id']
```

```python
np.mean(train['année'])
```

```python
for i in numerical_cols:
  train[i].fillna(np.mean(train[i]), inplace =True)
```

```python
train.isnull().sum()
```

**Observations:**

- We can see that we have successfully treated all the **numerical columns** who have missing values.

```python
# Making a list for all of the categorical variables
categorical_cols = []

# Picking up all of the categorcial columns one after another and diumping it into the categorical_cols list
for i in train.columns:
  if train[i].dtypes == 'object':
    categorical_cols.append(i)
```

```python
categorical_cols
```

```python
# Replacing all of the missing values of the categical_columns with their respective mode

for i in categorical_cols:
  train[i] = train[i].replace(np.nan ,train[i].mode()[0])
```

```python
# Missing values check
train.isnull().sum()
```

**Observations:**

- We can see that we have **successfully cleaned** all of the **missing values** from our dataset.

### **6.1 Univariate Analysis**

<a name = Section61></a>
**<h4>Question:** What insights can be drawn from all of the **continious features** separately?

```python
from tqdm.notebook import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


# Separating numerical and categrical columns
categorical_cols,numerical_cols = [],[]
for i in tqdm(train.columns):
  if train[i].dtypes !='object':
    numerical_cols.append(i)
  if train[i].dtypes =='object':
    categorical_cols.append(i)


for i in tqdm(numerical_cols):
  plt.figure(figsize = [15,3])
  sns.set(style="darkgrid")
  plt.subplot(1, 2, 1)
  sns.boxplot(data=train,x=i,color='orange')
  plt.subplot(1, 2, 2)
  sns.distplot(a=train[i],color='darkblue')
  plt.xticks(rotation=45)
  plt.grid()
  plt.show()
```

**Observations:**

- There are too many outliers in the **price** column and it is sensetive to outliers as well.

- **price** shows us a **right skewed distribution**.

- There are too many outliers in the **annee** column and it is sensetive to outliers as well.
